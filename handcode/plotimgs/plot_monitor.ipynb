```python
from matplotlib import pyplot as plt
from scipy.interpolate import make_interp_spline
from matplotlib.backends.backend_pdf import PdfPages
from scipy.io import savemat
import numpy as np
import pickle
import sys
import seaborn as sns


def load_pickle(file_path):
  with open(file_path, 'rb') as f:
    return pickle.load(f)
  return 1
mt, labels = load_pickle("./monitor_acm.pickle")
```

```python
h = mt['h']
hn = np.array(h)
hn = np.array(hn.flat)
print(hn.shape)
sns.distplot(hn)
savemat("./monitor-h.mat", {"monitor-h": hn})
```

```python
num_labels = len(set(labels))
```

```python
T = len(mt['h'])
t_ind = 10
img_ind = 12
nr_ind = 2
lis_votages = []
font1 = {'family' : 'Times New Roman', 'weight' : 'bold','size'   : 14}
plt.figure(figsize=(10, 2.5))
print("label:", labels[12])
f, ax = plt.subplots(1,3, figsize=(10,2.5))
for nr_ind in range(3):
    vtgs=[]
    for t in range(T):
        vtgs.append(mt['h'][t][img_ind][nr_ind])
    y,x = np.array(vtgs), np.arange(0, len(vtgs))
    
    lis_votages.append(y)
    
    xnew = np.linspace(x.min(), x.max(), 300) 
    spl = make_interp_spline(x, y, k=3)
    ysmooth= spl(xnew)
    
    ax[nr_ind].plot(xnew, ysmooth,"b", linewidth=1.3)
    ax[nr_ind].hlines(0.6, 0, 100,color="r",linestyles="dashed", label="threshold=0.6", linewidth=2)
    if nr_ind == 0: ax[nr_ind].set_ylabel("voltage", font1)
    ax[nr_ind].set_xlabel("T", font1)
    ax[nr_ind].legend()
    tlabels = ax[nr_ind].get_xticklabels() + ax[nr_ind].get_yticklabels()
    [label.set_fontweight('bold') for label in tlabels]
    
    if nr_ind == 0: ax[nr_ind].legend(loc=4)
    if nr_ind == 1: ax[nr_ind].legend(loc=7)
        
lis_votages = np.array(lis_votages)
savemat("./votages.mat", {"votages": lis_votages})
        
plt.legend()
plt.savefig('monitor-h.pdf', dpi=200, bbox_inches='tight')
# plt.subplots_adjust(wspace=2, hspace=None)
plt.show()
```
