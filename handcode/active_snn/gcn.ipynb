```python
import graphgallery as gg
from graphgallery import functional as gf
from graphgallery.datasets import Planetoid, NPZDataset
data = Planetoid('cora', root="~/datasets/datafromgg", verbose=False)
graph = data.graph
splits = data.split_nodes()
print("loaded")
```

```python
import numpy as np
import random
runs =100 
scores = []
for run in range(runs):
    train_nodes = splits.train_nodes
    val_nodes = splits.val_nodes
    # train_nodes = train_nodes[:35]
    # train_nodes = np.hstack([train_nodes,val_nodes])
    print(len(train_nodes))
    random_tr_ind = list(range(len(train_nodes)))
    random.shuffle(random_tr_ind)
    num_per_class = 20
    sample_result =[]
    classes = np.zeros(7)
    for i in random_tr_ind:
        if classes[graph.y[i]]<num_per_class:
            sample_result.append(i)
            classes[graph.y[i]]+=1
    sample_result = np.array(sample_result)
    print(sample_result)
    # val = list(set(train_nodes)-set(sample_result))
    # print(np.array(val))
    trainer = gg.gallery.nodeclas.GCN(seed=None, device="gpu").make_data(graph).build()
    his = trainer.fit(sample_result,verbose=1, dropout=0,epochs=200,ModelCheckpoint=dict(enabled=False))
    score = trainer.evaluate(splits.test_nodes)
    print("type(score)",type(score))
    scores.append(score.accuracy)
print("accuracy:",np.mean(scores), "std:", np.std(scores) )
```
